```python
#upload all station data
df = pd.read_csv("asos.csv", low_memory=False)
```

```python
#check features
df.describe()
```

```python
#drop features with low count

df.drop(['mslp', 'ice_accretion_1hr', 'ice_accretion_3hr', 'ice_accretion_6hr', 'snowdepth', 'metar', 'peak_wind_gust', 'peak_wind_drct'], axis=1, inplace=True)
```

```python
#fix date time

df['date'] = pd.to_datetime(df['valid'])
df.drop(['valid'], axis=1, inplace=True)
```

```python
#seperate into stations
print(df['station'].unique())

#bjc
df_bjc = df[df['station']=='BJC']
df_bjc.reindex()
df_bjc.drop(['station'], axis=1, inplace=True)

#bdu
df_bdu = df[df['station']=='BDU']
df_bdu.drop(['station'], axis=1, inplace=True)

#lmo
df_lmo = df[df['station']=='LMO']
df_lmo.drop(['station'], axis=1, inplace=True)

#gnb
df_gnb = df[df['station']=='GNB']
df_gnb.drop(['station'], axis=1, inplace=True)

#c99
df_c99 = df[df['station']=='C99']
df_c99.drop(['station'], axis=1, inplace=True)

del df
```

```python
#lets take a look at when strong winds are happening

count = (df_bjc['sknt'] > 35).sum()
perc = count/len(df_bjc) * 100
print('bjc', ':',count, perc)

count = (df_bdu['sknt'] > 35).sum()
perc = count/len(df_bdu) * 100
print('bdu', ':',count, perc)

count = (df_lmo['sknt'] > 35).sum()
perc = count/len(df_lmo) * 100
print('lmo', ':',count, perc)

count = (df_gnb['sknt'] > 35 or df_gnb['gust'] > 35).sum()
perc = count/len(df_gnb) * 100
print('gnb', ':',count, perc)

count = (df_c99['sknt'] > 35).sum()
perc = count/len(df_c99) * 100
print('c99', ':',count, perc)
```

```python
df_bd = pd.read_csv("BDU.csv")
print(df_bd.columns)
df_bd.iloc[0]
```

```python
num_col = ['tmpf', 'dwpf', 'relh', 'drct', 'sknt', 'alti', 'vsby', 'skyl1', 'skyl2', 'skyl3', 'feel']

for col in num_col:
    df_bd[col] = pd.to_numeric(df_bd[col], errors='coerce')
```

```python
col_to_keep = [col for col in df_bd.columns if df_bd[col].nunique() < 2]

df_bd.drop(col_to_keep, axis=1, inplace=True)
df_bd.drop(['metar'], axis=1, inplace=True)
```

```python
df_bd["valid"] = pd.to_datetime(df_bd["valid"])  
df_bd['valid'] = df_bd['valid'].dt.date
```

```python
agg_dict = {col: ['mean', 'min', 'max'] for col in num_col}
```

```python
df = df_bd.groupby('valid').agg(agg_dict)
df.columns = ['_'.join(col) for col in df.columns]
```
