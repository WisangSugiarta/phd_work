### Kuramoto-Sivashinsky Equations

$$\frac{\partial u}{\partial t} + u \frac{\partial u}{\partial x} + \frac{\partial^2 u}{\partial x^2} + v\frac{\partial^4 u}{\partial x^4} = 0$$

This PDE is a simple turbulence model.

```python
L = 32 * np.pi             
N = 256                 
x = np.linspace(0, L, N, endpoint=False)
dx = L / N

dt = 0.0001                 
T = 150       
n_steps = int(T / dt)

# Fourier wave numbers
k = np.fft.fftfreq(N, d=dx) * 2 * np.pi
ik = 1j * k
k2 = k ** 2
k4 = k ** 4

# Initial condition: smooth bump + noise
u = np.cos(x / 16) * (1 + np.sin(x / 16)) + 0.01 * np.random.randn(N)

# Right-hand side using spectral derivatives
def ks_rhs(u):
    u_hat = np.fft.fft(u)
    u_x = np.fft.ifft(ik * u_hat).real
    u_xx = np.fft.ifft(-k2 * u_hat).real
    u_xxxx = np.fft.ifft(k4 * u_hat).real
    return -u * u_x - u_xx - u_xxxx

# RK4 step
def rk4_step(u, dt):
    k1 = ks_rhs(u)
    k2 = ks_rhs(u + 0.5 * dt * k1)
    k3 = ks_rhs(u + 0.5 * dt * k2)
    k4 = ks_rhs(u + dt * k3)
    return u + (dt / 6) * (k1 + 2*k2 + 2*k3 + k4)

# Save solution snapshots
save_every = 5
saved_u = []
saved_t = []

for step in range(n_steps):
    u = rk4_step(u, dt)
    if step % save_every == 0:
        saved_u.append(u.copy())
        saved_t.append(step * dt)

# Convert to array for plotting
saved_u = np.array(saved_u)
saved_t = np.array(saved_t)

# Plot space-time diagram
plt.figure(figsize=(10, 5))
plt.imshow(saved_u, aspect='auto', extent=[0, L, saved_t[-1], saved_t[0]],
           cmap='plasma')
plt.colorbar(label='u(x, t)')
plt.xlabel('x')
plt.ylabel('time')
plt.title('Kuramoto–Sivashinsky Simulation')
plt.tight_layout()
plt.show()
```

### local linear factor

```python
def local_linear_forecast(X, Y, query_idx, k=10):
    nbrs = NearestNeighbors(n_neighbors=k+1).fit(X)
    distances, indices = nbrs.kneighbors([X[query_idx]])
    
    neighbor_idx = indices[0][1:]
    
    X_neighbors = X[neighbor_idx]
    Y_neighbors = Y[neighbor_idx]
    
    model = LinearRegression().fit(X_neighbors, Y_neighbors)
    pred = model.predict([X[query_idx]])
    return pred[0]
```

### Method of Analogues

```python
from sklearn.neighbors import NearestNeighbors

def method_of_analogues(U, n_ahead=1, k=5):
    """
    U: array of shape (n_snapshots, n_grid) -- each row is a KS state
    n_ahead: how many steps ahead to forecast
    k: number of neighbors to average over
    """
    U_train = U[:-n_ahead]
    U_target = U[n_ahead:]

    def forecast(u_now):
        nbrs = NearestNeighbors(n_neighbors=k).fit(U_train)
        dists, idxs = nbrs.kneighbors([u_now])
        prediction = np.mean(U_target[idxs[0]], axis=0)
        return prediction

    return forecast
```

```python
%%time

tau = 100
embed_dim = 4
k = 1
train_frac = 0.7

u_t = (u_t - np.mean(u_t)) / np.std(u_t)

def method_of_analogues(X_train, Y_train, x_query, k=1):
    nn = NearestNeighbors(n_neighbors=k).fit(X_train)
    _, idxs = nn.kneighbors([x_query])
    return np.mean(Y_train[idxs[0]])

def forecast_all_moa(X_train, Y_train, X_test, k=1):
    return np.array([method_of_analogues(X_train, Y_train, x, k) for x in X_test])

X, Y = create_embedding(u_t, tau, m=embed_dim)

X_train, Y_train, X_test, Y_test = split_train_test(X, Y, frac=train_frac)
Y_pred = forecast_all_moa(X_train, Y_train, X_test, k=k)

rmse = np.sqrt(mean_squared_error(Y_test, Y_pred))
mae = mean_absolute_error(Y_test, Y_pred)
r2 = r2_score(Y_test, Y_pred)

print(f"RMSE: {rmse:.4f}")
print(f"MAE:  {mae:.4f}")
print(f"R²:   {r2:.4f}")

plt.figure(figsize=(10, 4))
plt.plot(Y_test[:2000], label='True')
plt.plot(Y_pred[:2000], '--', label='Predicted')
plt.xlabel('Time step (index)')
plt.ylabel(f'u(t + {embed_dim}τ)')
plt.title('MoA Prediction vs Ground Truth (Long-Term)')
plt.legend()
plt.tight_layout()
plt.show()
```

### Reservoir

```python
%%time

# === 1. Normalize and embed time series ===
u_t = (u_t - np.mean(u_t)) / np.std(u_t)

def create_embedding(u_t, tau, m):
    N = len(u_t) - tau * (m + 1)
    X = np.column_stack([u_t[i * tau : i * tau + N] for i in range(m)])
    Y = u_t[m * tau : m * tau + N].reshape(-1, 1)  # <<< Make Y shape (N, 1)
    return X, Y

tau = 100
embed_dim = 4
train_frac = 0.8

X, Y = create_embedding(u_t, tau, embed_dim)
split = int(len(X) * train_frac)
X_train, Y_train = X[:split], Y[:split]
X_test, Y_test = X[split:], Y[split:]

# === 2. Build reservoir and readout ===
res = Reservoir(units=500, lr=0.3, sr=1.25, seed=42)
readout = Ridge(ridge=1e-6)

# === 3. Train ===
states_train = res.run(X_train)
readout.fit(states_train, Y_train)  # <<< shapes: (N, 500), (N, 1)

# === 4. Predict ===
states_test = res.run(X_test)
Y_pred = readout.run(states_test)  # <<< shape: (N, 1)

# === 5. Metrics ===
rmse = np.sqrt(mean_squared_error(Y_test, Y_pred))
mae = mean_absolute_error(Y_test, Y_pred)
r2 = r2_score(Y_test, Y_pred)
print(f"RMSE: {rmse:.4f}, MAE: {mae:.4f}, R²: {r2:.4f}")

# === 6. Plot ===
plt.figure(figsize=(10, 4))
plt.plot(Y_test[:2000], label='True')
plt.plot(Y_pred[:2000], '--', label='RC Predicted')
plt.xlabel('Time step')
plt.ylabel(f'u(t + {embed_dim}τ)')
plt.title('Reservoir Computing Prediction of KS')
plt.legend()
plt.tight_layout()
plt.show()
```
